````python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx 
import time
````

````python
import networkx as nx 

G = nx.read_weighted_edgelist('edge_list.txt', delimiter =" ") 

population = { 
        'T1' : 100, 
        'T2' : 100, 
        'T3' : 100, 
        'T4' : 100, 
        'T5' : 100,
        'T6' : 100,
        'T7' : 100,
        'T8' : 100
        } 
  
# We have to set the population attribute for each of the 14 nodes 
for i in list(G.nodes()): 
    G.nodes[i]['population'] = population[i] 
  
nx.draw_networkx(G, with_label = True) 
# This line allows us to visualize the Graph
````

## Display 10 graphs and change the weight of edges

#### Goal : Display the same graph and replace the older by the newer

````python
count = 0
while count < 10 :
    for i in list(G.nodes()): 
        G.nodes[i]['population'] = population[i] 

    # fixing the size of the figure 
    plt.figure(figsize =(10, 7))

    node_color = [G.degree(v) for v in G]
    # node colour is a list of degrees of nodes

    node_size = [100*count+1.5 * nx.get_node_attributes(G, 'population')[v] for v in G] 
    # size of node is a list of population of cities

    edge_width = [0.0015 * G[u][v]['weight'] for u, v in G.edges()]
    # width of edge is a list of weight of edges 

    nx.draw_networkx(G, node_size = node_size,  
                     node_color = node_color, alpha = 0.7, 
                     with_labels = True, width = edge_width, 
                     edge_color ='.4', cmap = plt.cm.Blues) 

    plt.axis('off')
    plt.tight_layout(); 
    count+=1
    time.sleep(3)
````

````python
import numpy as np
import networkx as nx
from random import *

## création d'une matrice d'adjacense random
def generateMatrix():
    first = list()
    for i in range(4):
        second = list()
        for j in range(4):
            # add to first list
            n = randint(0,1)
            second.append(n)
        # add to second list
        first.append(second)
    print(first)
    return first

# Update : poids des sommets.
## Etape 1 : trouver comment ajuster les poids des sommets sur une matrice d'adjacence
## Etape 2 : update le poids des sommets

A=np.matrix(generateMatrix())
G=nx.from_numpy_matrix(A)
nx.draw_networkx(G, with_label = True) 
````

````python
DG = nx.DiGraph()
# dict-of-dict-of-attribute
adj = {1: {2: 1.3, 3: 0.7}, 2: {1: 1.4}, 3: {1: 0.7}}
e = [(u, v, {'weight': d}) for u, nbrs in adj.items()
        for v, d in nbrs.items()]
print(e)
DG.update(edges=e, nodes=adj)
nx.draw_networkx(DG, with_label = True) 
````

````python
import sys, networkx as nx, matplotlib.pyplot as plt

# Create a list of 4 nodes numbered [0, 3]
nodes = range(4)
node_sizes = []
labels = {}
for n in nodes:
        node_sizes.append( 100 * n )
        labels[n] = 100 * n

# Node sizes: [0, 100, 200, 300]

# Connect each node to its successor
edges = [ (i, i+1) for i in range(len(nodes)-1) ]

# Create the graph and draw it with the node labels
g = nx.Graph()
g.add_nodes_from(nodes)
g.add_edges_from(edges)

nx.draw_networkx(g, node_size = node_sizes, labels=labels, with_labels=True)    
plt.show()
````

# Use Pyvis : dynamic network librairy

````python
from pyvis.network import Network
import networkx as nx

## Display
G = Network(height="750px", width="100%", bgcolor="#222222", font_color="white")

## Create the networkx graph
g = nx.Graph()
# Create a list of 4 nodes numbered [0, 3]
nodes = range(4)
node_sizes = []
labels = {}
for n in nodes:
        node_sizes.append( 100 * n )
        labels[n] = 100 * n

# Node sizes: [0, 100, 200, 300]

# Connect each node to its successor
edges = [ (i, i+1) for i in range(len(nodes)-1) ]

# Create the graph and draw it with the node labels

g.add_nodes_from(nodes)
g.add_edges_from(edges)

G.from_nx(g)
G.show("simu/mygraph.html")
````

````python
from pyvis.network import Network

## Display

g = Network(height="750px", width="100%", bgcolor="#222222", font_color="white")
g.barnes_hut()
## Create the networkx graph
# Create a list of 4 nodes numbered [0, 3]
nodes = range(4)
node_sizes = []
labels = {}
for n in nodes:
        node_sizes.append( 100 * n )
        labels[n] = 100 * n
        g.add_node(n)

# Node sizes: [0, 100, 200, 300]

# Connect each node to its successor
for i in range(len(nodes)-1):
    g.add_edge(i, i+1, weight=i*3/2)

# Create the graph and draw it with the node labels



g.show("simu/customGrap.html")

````
